## Step 4　ナレーションリストを編集する

`NARRATIONS` リストの `id` と `text` を書き換えてください。

- `id`：MP3 のファイル名になります（**英数字・アンダースコアのみ** / 日本語不可）
- `text`：読み上げるテキスト（1 リクエスト最大 5,000 バイト ≈ 日本語約 1,600 文字）

▶ 編集が終わったらセルを実行してください。登録件数が表示されます。

```python
from typing import TypedDict


class Narration(TypedDict):
    id: str    # MP3 のファイル名（英数字・アンダースコアのみ）
    text: str  # 読み上げるテキスト


NARRATIONS: list[Narration] = [
    {'id': '01_intro',    'text': 'この動画では〇〇について解説します。'},
    {'id': '02_chapter1', 'text': '第一章。まず、〇〇の基本から見ていきましょう。'},
    # 必要な行数だけ追加してください
]

print(f'✅ ナレーション {len(NARRATIONS)} 件を登録しました')
for item in NARRATIONS:
    preview = item['text'][:30] + ('…' if len(item['text']) > 30 else '')
    print(f'   {item["id"]:<20}  {preview}')
```

## Step 5　TTS 関数の定義

▶ セルを実行するだけで OK です。**「✅ 関数定義完了」** と表示されれば次へ進めます。

> ℹ️ コードを読まなくても大丈夫ですが、仕組みが気になる方へ：
> - `build_client()` ：GCP の認証情報を使って API 接続を作ります
> - `synthesize()` ：テキスト 1 件を受け取って MP3 データ（バイト列）を返します

```python
import google.auth
from pathlib import Path
from google.cloud import texttospeech
from google.api_core.exceptions import GoogleAPICallError


def build_client() -> texttospeech.TextToSpeechClient:
    """GCP の認証情報を使って TTS クライアントを作成する。"""
    credentials, _ = google.auth.default(
        scopes=['https://www.googleapis.com/auth/cloud-platform']
    )
    # 課金・クォータの対象プロジェクトを明示的に指定
    credentials = credentials.with_quota_project(PROJECT_ID)
    return texttospeech.TextToSpeechClient(credentials=credentials)


def synthesize(client: texttospeech.TextToSpeechClient, text: str) -> bytes:
    """テキストを MP3 音声データ（バイト列）に変換して返す。"""
    response = client.synthesize_speech(
        input=texttospeech.SynthesisInput(text=text),
        voice=texttospeech.VoiceSelectionParams(
            language_code='ja-JP',
            name=TTS_VOICE,
        ),
        audio_config=texttospeech.AudioConfig(
            audio_encoding=texttospeech.AudioEncoding.MP3,
            speaking_rate=TTS_SPEAKING_RATE,
            pitch=TTS_PITCH,
        ),
    )
    return response.audio_content


print('✅ 関数定義完了')
```

## Step 7　一括生成

`NARRATIONS` リストの全件を順番に生成して保存します。

▶ セルを実行すると各行の処理結果が表示されます。

| ログ | 意味 |
|------|------|
| `[OK]` | 生成成功 |
| `[SKIP]` | 同名ファイルが存在したためスキップ（`OVERWRITE = False` の場合） |
| `[ERROR]` | 生成失敗（エラー内容を確認してください） |

> ℹ️ `OVERWRITE = True` にすると全件上書きします。
> 特定ファイルだけ再生成したい場合は `/content/output/` から対象ファイルを削除してから実行してください。

```python
import sys
from pathlib import Path
from google.api_core.exceptions import GoogleAPICallError

if PROJECT_ID == 'YOUR_PROJECT_ID':
    print('⚠️  PROJECT_ID が設定されていません。Step 3 に戻って入力してください。')
else:
    output_path = Path(OUTPUT_DIR)
    output_path.mkdir(parents=True, exist_ok=True)
    client = build_client()

    generated = skipped = errors = 0

    for item in NARRATIONS:
        item_id = item['id']
        out_file = output_path / f"{item_id}.mp3"

        if not OVERWRITE and out_file.exists():
            print(f'[SKIP]  {item_id}')
            skipped += 1
            continue

        try:
            audio_data = synthesize(client, item['text'])
            out_file.write_bytes(audio_data)
            print(f'[OK]    {item_id}')
            generated += 1
        except GoogleAPICallError as e:
            print(f'[ERROR] {item_id}: {e.message}', file=sys.stderr)
            errors += 1
        except Exception as e:
            print(f'[ERROR] {item_id}: {e}', file=sys.stderr)
            errors += 1

    print()
    print(f'===== 完了: 生成 {generated}  スキップ {skipped}  エラー {errors} =====')
    if errors > 0:
        print('⚠️  エラーがあります。上のログを確認してください。')
    elif generated > 0:
        print('✅ 生成完了！ Step 8 でダウンロードしてください。')
```

## Step 8　ZIP でまとめてダウンロード

生成した MP3 ファイルを 1 つの ZIP にまとめてダウンロードします。

▶ セルを実行するとブラウザのダウンロードが自動で始まります。

> ℹ️ ダウンロードファイル名：`narration_output.zip`
> ZIP を展開すると `01_intro.mp3` のように `id` をファイル名とした MP3 ファイルが入っています。

```python
import shutil
from google.colab import files

zip_base = '/content/narration_output'
shutil.make_archive(zip_base, 'zip', OUTPUT_DIR)
files.download(f'{zip_base}.zip')
print('✅ ダウンロード開始')
```
