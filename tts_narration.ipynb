## Step 4　ナレーションリストを CSV で読み込む

▶ セルを実行すると「**CSV を選択**」ボタンが表示されます。ボタンをクリックして CSV ファイルをアップロードしてください。

### CSV フォーマット

| 列名 | 説明 |
|------|------|
| `id` | MP3 のファイル名（**英数字・アンダースコアのみ** / 日本語不可） |
| `text` | 読み上げるテキスト（1 リクエスト最大 5,000 バイト ≈ 日本語約 1,600 文字） |

**例（narrations.csv）:**
```
id,text
01_intro,この動画では〇〇について解説します。
02_chapter1,第一章。まず、〇〇の基本から見ていきましょう。
03_outro,ご視聴ありがとうございました。
```

> ℹ️ **UTF-8・Shift-JIS（Excel 保存）どちらでも読み込めます。**
> 読み込み後にプレビューが表示されるので、文字化けがないか確認してください。

````python
import csv
import io
from typing import TypedDict

import ipywidgets as widgets
from IPython.display import display, clear_output


class Narration(TypedDict):
    id: str    # MP3 のファイル名（英数字・アンダースコアのみ）
    text: str  # 読み上げるテキスト


_ENCODINGS = ("utf-8-sig", "utf-8", "cp932", "shift_jis")


def _decode_upload(upload_value) -> bytes:
    """ipywidgets 7 / 8 両対応。アップロード内容を bytes で返す。"""
    file_info = (
        next(iter(upload_value.values()))
        if isinstance(upload_value, dict)
        else upload_value[0]
    )
    content = file_info["content"]
    return bytes(content) if isinstance(content, memoryview) else content


def _detect_and_decode(raw: bytes) -> tuple[str, str]:
    """エンコードを順番に試し、(エンコード名, デコード済み文字列) を返す。"""
    for encoding in _ENCODINGS:
        try:
            return encoding, raw.decode(encoding)
        except UnicodeDecodeError:
            continue
    raise UnicodeDecodeError("utf-8", raw, 0, len(raw), "対応エンコードで読み取れませんでした")


def _parse_csv(raw: bytes) -> tuple[str, list[Narration]]:
    """CSV を解析し (検出エンコード, ナレーションリスト) を返す。"""
    encoding, text = _detect_and_decode(raw)
    reader = csv.DictReader(io.StringIO(text))
    items = [
        Narration(id=row["id"].strip(), text=row["text"].strip())
        for row in reader
    ]
    return encoding, items


class _NarrationUploader:
    def __init__(self, target: list[Narration]) -> None:
        self._target = target
        self._widget = widgets.FileUpload(
            accept=".csv", multiple=False, description="CSV を選択"
        )
        self._output = widgets.Output()
        self._widget.observe(self._on_upload, names="value")

    def _on_upload(self, change: dict) -> None:
        with self._output:
            clear_output()
            try:
                raw = _decode_upload(change["new"])
                encoding, items = _parse_csv(raw)
                self._target.clear()
                self._target.extend(items)

                print(f"✅ {len(self._target)} 件読み込みました（エンコード: {encoding}）")
                print("─" * 60)
                print("【プレビュー】文字化けがないか確認してください")
                print("─" * 60)
                for item in self._target:
                    preview = item["text"][:40] + ("…" if len(item["text"]) > 40 else "")
                    print(f"  [{item['id']}] {preview}")
            except (KeyError, csv.Error) as e:
                print(f"❌ CSV フォーマットエラー: {e}")
                print("   列名 id と text が含まれているか確認してください。")
            except UnicodeDecodeError as e:
                print(f"❌ 文字コードエラー: {e}")
                print("   UTF-8 または Shift-JIS で保存したファイルを選択してください。")

    def show(self) -> None:
        display(self._widget, self._output)


NARRATIONS: list[Narration] = []
_uploader = _NarrationUploader(NARRATIONS)
_uploader.show()
````

## Step 5　TTS 関数の定義

▶ セルを実行するだけで OK です。**「✅ 関数定義完了」** と表示されれば次へ進めます。

> ℹ️ コードを読まなくても大丈夫ですが、仕組みが気になる方へ：
> - `build_client()` ：GCP の認証情報を使って API 接続を作ります
> - `synthesize()` ：テキスト 1 件を受け取って MP3 データ（バイト列）を返します

````python
import google.auth
from pathlib import Path
from google.cloud import texttospeech
from google.api_core.exceptions import GoogleAPICallError


def build_client() -> texttospeech.TextToSpeechClient:
    """GCP の認証情報を使って TTS クライアントを作成する。"""
    credentials, _ = google.auth.default(
        scopes=['https://www.googleapis.com/auth/cloud-platform']
    )
    # 課金・クォータの対象プロジェクトを明示的に指定
    credentials = credentials.with_quota_project(PROJECT_ID)
    return texttospeech.TextToSpeechClient(credentials=credentials)


def synthesize(client: texttospeech.TextToSpeechClient, text: str) -> bytes:
    """テキストを MP3 音声データ（バイト列）に変換して返す。"""
    response = client.synthesize_speech(
        input=texttospeech.SynthesisInput(text=text),
        voice=texttospeech.VoiceSelectionParams(
            language_code='ja-JP',
            name=TTS_VOICE,
        ),
        audio_config=texttospeech.AudioConfig(
            audio_encoding=texttospeech.AudioEncoding.MP3,
            speaking_rate=TTS_SPEAKING_RATE,
            pitch=TTS_PITCH,
            sample_rate_hertz=24000,
        ),
    )
    return response.audio_content


print('✅ 関数定義完了')
````

## Step 7　一括生成

`NARRATIONS` リストの全件を順番に生成して保存します。

▶ セルを実行すると各行の処理結果が表示されます。

| ログ | 意味 |
|------|------|
| `[OK]` | 生成成功 |
| `[SKIP]` | 同名ファイルが存在したためスキップ（`OVERWRITE = False` の場合） |
| `[ERROR]` | 生成失敗（エラー内容を確認してください） |

> ℹ️ `OVERWRITE = True` にすると全件上書きします。
> 特定ファイルだけ再生成したい場合は `/content/output/` から対象ファイルを削除してから実行してください。

````python
import sys
from pathlib import Path
from google.api_core.exceptions import GoogleAPICallError

if PROJECT_ID == 'YOUR_PROJECT_ID':
    print('⚠️  PROJECT_ID が設定されていません。Step 3 に戻って入力してください。')
else:
    output_path = Path(OUTPUT_DIR)
    output_path.mkdir(parents=True, exist_ok=True)
    client = build_client()

    generated = skipped = errors = 0

    for item in NARRATIONS:
        item_id = item['id']
        out_file = output_path / f"{item_id}.mp3"

        if not OVERWRITE and out_file.exists():
            print(f'[SKIP]  {item_id}')
            skipped += 1
            continue

        try:
            audio_data = synthesize(client, item['text'])
            out_file.write_bytes(audio_data)
            print(f'[OK]    {item_id}')
            generated += 1
        except GoogleAPICallError as e:
            print(f'[ERROR] {item_id}: {e.message}', file=sys.stderr)
            errors += 1
        except Exception as e:
            print(f'[ERROR] {item_id}: {e}', file=sys.stderr)
            errors += 1

    print()
    print(f'===== 完了: 生成 {generated}  スキップ {skipped}  エラー {errors} =====')
    if errors > 0:
        print('⚠️  エラーがあります。上のログを確認してください。')
    elif generated > 0:
        print('✅ 生成完了！ Step 8 でダウンロードしてください。')
````
